# **Introdução**

O ano de 2020 começou para nós brasileiros e para a maioria do mundo como qualquer ano: muitas expectativas e promessas de realização profissionale pessoal.

O que ninguém imaginou é que um tímido vírus, que surgiu na cidade de Wuhan, localizada na Província central chinesa de Hubei, iria se alastrar para o mundo inteiro e causar um efeito globalizado indesejado: causar uma pandemia.

Já no ínicio de 2020, no primeiro trimestre fomos bombardeados por notícias mundiais de diversos países apresentando um rápido crescimento dos casos de contaminação pelo vírus que foi chamado de Sars-CoV-2 (ou COVID-19 de forma simplifada - **co**rona **vi**rus **d**isease-19 (ano de surgimento). 

E com o Brasil não foi diferente. Em Fevereiro de 2020, o país pós-carnaval começou a apresentar alguns casos de COVID-19, e, em Março, com o súbito aumento o Brasil entrou na lista dos países que estavam fazendo quarentena e *lockdown* das suas atividades: a regra é(era) clara, todos em casa e minimizar a exposição com o meio externo e contato com terceiros.

Aos poucos, o mundo parado em casa e aprendendo a fazer todas as atividades via infindáveis reuniões online, foi descobrindo mais sobre o vírus, suas formas de contaminação, seus sintomas mais comuns, os mais graves, como evitar exposição e como minimizar o risco de contrar o vírus.

Em meio a um mar de sintomas que atingem diferentes espectros de gravidade, encontramos um quadro de saúde preocupante que é causado pelo COVID-19: a Síndrome respiratória aguda grave (SRAG), que é uma condição crítica de saúde que necessita de hospitalização, ou seja, é uma condição grave agravada e gerada pelo COVID-19 em algumas pessoas.


Como será que foi a evolução dessa doença no Brasil? Será que ela impactou os casos de SRAG que o país vem acompanhando desde antes do surgimento do COVID-19? Qual é o impacto da contaminação nas diferentes regiões do Brasil?

```
#@title Gráfico Modelo Prophet - Treinamento com dados do ano de 2020 inteiro
prophet_br_casos_novos_treinado.plot(previsao_prophet_br_casos_novos,plot_cap=True, xlabel = 'Data', ylabel = 'Casos Novos COVID-19',figsize=(18,8));
```

O modelo Prophet trouxe uma previsão de aumento de casos novos por dia, com valores próximos de 55 mil casos por dia para o mês de fevereiro, contudo com um intervalo de dados que variam 10 mil casos para mais e para menos, aproximadamente.
